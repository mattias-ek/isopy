```python
import isopy
import numpy as np
import matplotlib.pyplot as plt
import warnings
import matplotlib as mpl

cc_val = isopy.array('Mo.xlsx', sheetname='CC_val')
cc_pm  = isopy.array('Mo.xlsx', sheetname='CC_pm')

nc_val = isopy.array('Mo.xlsx', sheetname='NC_val')
nc_pm  = isopy.array('Mo.xlsx', sheetname='NC_pm')

cai_val = isopy.array('Mo.xlsx', sheetname='CAI_val')
cai_pm  = isopy.array('Mo.xlsx', sheetname='CAI_pm')
```

## scatter

```python
isopy.plt.scatter(cc_val['mo94'], cc_val['mo95'], cc_pm['mo94'], cc_pm['mo95'], label = 'CC')
isopy.plt.scatter(nc_val['mo94'], nc_val['mo95'], nc_pm['mo94'], nc_pm['mo95'], label = 'NC')
isopy.plt.scatter(cai_val['mo94'], cai_val['mo95'], cai_pm['mo94'], cai_pm['mo95'], label = 'CAI')
```

```python
isopy.plt.update_figure(height=4, width=4)
isopy.plt.update_axes(xlabel='$\epsilon{}^{94}$Mo', ylabel='$\epsilon{}^{95}$Mo', 
                      xlim=(-1, 5.5), ylim=(-1, 4), cross=0)

isopy.plt.scatter(cc_val['mo94'], cc_val['mo95'], cc_pm['mo94'], cc_pm['mo95'], label = 'CC')
isopy.plt.scatter(nc_val['mo94'], nc_val['mo95'], nc_pm['mo94'], nc_pm['mo95'], label = 'NC')
isopy.plt.scatter(cai_val['mo94'], cai_val['mo95'], cai_pm['mo94'], cai_pm['mo95'], label = 'CAI')

isopy.plt.legend(errorbars=False)
```

```python
kwargs = dict(axes_xlabel=dict(a='$\epsilon{}^{94}$Mo', b='$\epsilon{}^{100}$Mo'),
              axes_ylabel='$\epsilon{}^{95}$Mo', 
              axes_xlim=dict(a=(-1, 5.5), b=(-1, 3)),
              axes_ylim=(-1, 4), axes_cross = 0)

isopy.plt.create_subplots([['a', 'b']], row_height=4, column_width=4, **kwargs)

isopy.plt.a.scatter(cc_val['mo94'], cc_val['mo95'], cc_pm['mo94'], cc_pm['mo95'], label = 'CC')
isopy.plt.a.scatter(nc_val['mo94'], nc_val['mo95'], nc_pm['mo94'], nc_pm['mo95'], label = 'NC')
isopy.plt.a.scatter(cai_val['mo94'], cai_val['mo95'], cai_pm['mo94'], cai_pm['mo95'], label = 'CAI')

isopy.plt.b.scatter(cc_val['mo100'], cc_val['mo95'], cc_pm['mo100'], cc_pm['mo95'], label = 'CC')
isopy.plt.b.scatter(nc_val['mo100'], nc_val['mo95'], nc_pm['mo100'], nc_pm['mo95'], label = 'NC')
isopy.plt.b.scatter(cai_val['mo100'], cai_val['mo95'], cai_pm['mo100'], cai_pm['mo95'], label = 'CAI')
```

## regression

```python
isopy.plt.update_figure(height=6, width=6)
isopy.plt.update_axes(xlabel='$\epsilon{}^{94}$Mo', ylabel='$\epsilon{}^{95}$Mo', 
                      xlim=(-1, 5.5), ylim=(-1, 4), cross=0)

isopy.plt.scatter(nc_val['mo94'], nc_val['mo95'], nc_pm['mo94'], nc_pm['mo95'], 
                  label = 'NC', color='red', regression='york2')
isopy.plt.scatter(cc_val['mo94'], cc_val['mo95'], cc_pm['mo94'], cc_pm['mo95'], 
                  label = 'CC', color='blue', regression='york2')
isopy.plt.legend(errorbars=False)
```

```python
isopy.plt.update_figure(height=6, width=6)
isopy.plt.update_axes(xlabel='$\epsilon{}^{94}$Mo', ylabel='$\epsilon{}^{95}$Mo', 
                      xlim=(-1, 5.5), ylim=(-1, 4), cross=0)

isopy.plt.scatter(cc_val['mo94'], cc_val['mo95'], cc_pm['mo94'], cc_pm['mo95'], label = 'CC')
isopy.plt.scatter(nc_val['mo94'], nc_val['mo95'], nc_pm['mo94'], nc_pm['mo95'], label = 'NC')

isopy.plt.regression('linear')
isopy.plt.regression('york2', 'NC', color='red')

cc_regression = isopy.tb.yorkregress2(cc_val['mo94'], cc_val['mo95'], cc_pm['mo94'], cc_pm['mo95'])
isopy.plt.regression(cc_regression, color='blue')

isopy.plt.legend(fontsize='small')
```

## ystack & xstack 

```python
isopy.plt.ystack(cc_val['95mo'], cc_pm['95mo'], cval=isopy.mean, pmval=isopy.sd2, label='CC')
isopy.plt.ystack(nc_val['95mo'], nc_pm['95mo'], cval=isopy.mean, pmval=isopy.sd2, label='NC')
isopy.plt.update_axes(ylabel = '$\epsilon{}^{95}$Mo')
isopy.plt.legend()
```

```python
isopy.plt.ystack(cc_val['95mo'], cc_pm['95mo'], cval=isopy.mean, pmval=isopy.sd2, label='CC')
isopy.plt.ystack(nc_val['95mo'], nc_pm['95mo'], cval=isopy.mean, pmval=isopy.sd2, label='NC')
isopy.plt.legend()
isopy.plt.summary_ystack(cval=0, pmval=1)
```

```python
isopy.plt.xstack(nc_val, nc_pm, cval=isopy.nanmean, pmval=isopy.nansd2, label='NC')
isopy.plt.xstack(cc_val, cc_pm, cval=isopy.nanmean, pmval=isopy.nansd2, label='CC')
isopy.plt.legend()
```

```python
isopy.plt.xstack(nc_val, nc_pm, cval=isopy.nanmean, pmval=isopy.nansd2, label='NC', 
                 subplot_legend=True, summary_cval=0, summary_pmval=1, summary_combine_ticklabels=True)
isopy.plt.xstack(cc_val, cc_pm, cval=isopy.nanmean, pmval=isopy.nansd2, label='CC',
                 subplot_legend=True, summary_cval=0, summary_pmval=1, summary_combine_ticklabels=True)
isopy.plt.legend()
```
