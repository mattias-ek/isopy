```python
import isopy
import isopy.testing
import numpy as np
```

# test_100_1_text

```python
a = isopy.random(20, (1, 0.1), 'ru pd cd'.split(), seed = 46, dtype = [np.float64, np.float64, np.int8])
b = a.tabulate(include_row = True, include_dtype=True, nrows=12)
c = a.tabulate(row_names = [f'row {i+1}' for i in range(20)], floatfmt='.2f', intfmt='.1f')
d = a.tabulate(row_names=[f'row {i + 1}' for i in range(20)], nrows = 12,
                          floatfmt='f{:.2f}', intfmt='i{:.1f}')
e = a.tabulate(keys='pd ru')
f = a.tabulate(key_eq=['pd', 'ru'])
g = a.tabulate(keys='ru cd', key_eq=['pd', 'ru'])
h = a.tabulate('simple')
i = a.tabulate('markdown')
j = a.tabulate('markdown', include_objinfo = True)
k = a.tabulate('latex')
l = a.tabulate('latex', include_objinfo = True)
m = a.tabulate('html')
n = a.tabulate('html', include_objinfo = True)
o = a.to_refval()
```

## repr, str

```python
isopy.testing.generate_hash_test(
    a=a,
    b=(b, "a.tabulate(include_row = True, include_dtype=True, nrows=12)"),
    c=(c, "a.tabulate(row_names = [f'row {i+1}' for i in range(20)], floatfmt='.2f', intfmt='.1f')"),
    d=(d, "a.tabulate(row_names=[f'row {i + 1}' for i in range(20)], nrows = 12, \nfloatfmt='f{:.2f}', intfmt='i{:.1f}')"),
    e=(e, "a.tabulate(keys='pd ru')"),
    f=(f, "a.tabulate(key_eq=['pd', 'ru'])"),
    g=(g, "a.tabulate(keys='ru cd', key_eq=['pd', 'ru'])"),
    h=(h, "a.tabulate('simple')"),
    i=(i, "a.tabulate('markdown')"),
    j=(j, "a.tabulate('markdown', include_objinfo = True)"),
    k=(k, "a.tabulate('latex')"),
    l=(l, "a.tabulate('latex', include_objinfo = True)"),
    m=(m,"a.tabulate('html')"),
    n=(n,"a.tabulate('html', include_objinfo = True)"),
    o=(o,"a.to_refval()"),
    )
```

## html

```python
isopy.testing.generate_hash_test_html(str=True, repr=True, a=a, o=(o,"a.to_refval()"),)
```

```python
isopy.testing.generate_hash_test_html(
    b=(b, "a.tabulate(include_row = True, include_dtype=True, nrows=12)"),
    c=(c, "a.tabulate(row_names = [f'row {i+1}' for i in range(20)], floatfmt='.2f', intfmt='.1f')"),
    d=(d, "a.tabulate(row_names=[f'row {i + 1}' for i in range(20)], nrows = 12, \nfloatfmt='f{:.2f}', intfmt='i{:.1f}')"),
    e=(e, "a.tabulate(keys='pd ru')"),
    f=(f, "a.tabulate(key_eq=['pd', 'ru'])"),
    g=(g, "a.tabulate(keys='ru cd', key_eq=['pd', 'ru'])"),
    m=(m,"a.tabulate('html')"),
    n=(n,"a.tabulate('html', include_objinfo = True)"),
    o=(o,"a.to_refval()"),)
```

## markdown

```python
isopy.testing.generate_hash_test_markdown(
    i=(i, "a.tabulate('markdown')"),
    j=(j, "a.tabulate('markdown', include_objinfo = True)"))
```

## latex

```python
isopy.testing.generate_hash_test_latex(
    k=(k, "a.tabulate('latex')"),
    l=(l, "a.tabulate('latex', include_objinfo = True)"))
```

# test_1_1_text

```python
a = isopy.random(1, (1, 0.1), 'ru pd cd'.split(), seed=46, dtype = [np.float64, np.float64, np.int8])
b = a.tabulate(row_names=[f'row 1'])
c = a.tabulate(row_names=f'row 1')
```

## repr, str

```python
isopy.testing.generate_hash_test(
    a=a,
    b=(c, "a.tabulate(row_names=[f'row 1'])"),
    c=(c, "a.tabulate(row_names=f'row 1')"))
```

## html

```python
isopy.testing.generate_hash_test_html(str=True, repr=True, a=a)
```

```python
isopy.testing.generate_hash_test_html(
    a=a,
    b=(c, "a.tabulate(row_names=[f'row 1'])"),
    c=(c, "a.tabulate(row_names=f'row 1')"))
```

# test_1_0_text

```python
a = isopy.random(-1, (1, 0.1), 'ru pd cd'.split(), seed=46, dtype=[np.float64, np.float64, np.int8])
b = a.tabulate(row_names=[f'row 1'])
c = a.tabulate(row_names=f'row 1')
d = a.to_refval()
e = isopy.random(1, (1, 0.1), 'ru pd cd'.split(), seed=46, dtype=[np.float64, np.float64, np.int8])[0]
```

```python
isopy.testing.generate_hash_test_html(str=True, repr=True, 
    a=a, 
    d=(d,"a.to_refval()"), 
    e=(e,"isopy.random(1, (1, 0.1), 'ru pd cd'.split(), seed=46, dtype=[np.float64, np.float64, np.int8])[0]"))
```
